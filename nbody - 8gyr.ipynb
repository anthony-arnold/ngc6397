```python
from IPython.display import display, Math, Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = [12, 12]
mpl.rcParams['figure.dpi'] = 72

bound = np.loadtxt('8gyr/fort.31')
bound2 = np.loadtxt('8gyr/bound')


fig, ax = plt.subplots(ncols=1)
ax.set(
    title=r'Bound mass and stars',
    xlabel=r'T [Gyr]'
)

ax.grid()

ax.set_ylabel('ratio')
plt_1 = ax.plot(bound[:,1], bound[:,-1], color='red', label='Mass')
#ax.tick_params(axis ='y', labelcolor = 'red') 


#ax = ax.twinx()
ax.set_ylabel(r'N', color='blue')
plt_2 = ax.plot(bound[:,1], bound[:,2]/bound[0,2], color='blue', label='Stars')
#ax.tick_params(axis ='y', labelcolor = 'blue') 


fig, ax = plt.subplots(ncols=1)
ax.set(
    title=r'Bound mass and stars',
    xlabel=r'T [Gyr]'
)

'''
ax.set_ylabel(r'$M_{(t)} / M_0$', color='red')
plt_1 = ax.plot(bound[:,1], bound[:,-1], color='red', label='Mass')
ax.tick_params(axis ='y', labelcolor = 'red') 


ax = ax.twinx()
ax.set_ylabel(r'N', color='blue')
plt_2 = ax.plot(bound[:,1], bound[:,2], color='blue', label='Stars')
ax.tick_params(axis ='y', labelcolor = 'blue') 
'''

ax.set_ylabel('Ratio')

ax.grid()
plt_1 = ax.plot(bound2[:,0], bound2[:,2], color='red', label='Mass')
#ax.tick_params(axis ='y', labelcolor = 'red') 


#ax = ax.twinx()
#ax.set_ylabel(r'$N_{(t)} / N_0$', color='blue')
plt_2 = ax.plot(bound2[:,0], bound2[:,3], color='blue', label='Stars')
#ax.tick_params(axis ='y', labelcolor = 'blue') 

lns = plt_1 + plt_2
labels = [l.get_label() for l in lns]
plt.legend(lns, labels, loc=0)
```

```python
orbit = np.loadtxt('8gyr/NGC6397.orb_irr', skiprows=1)
nbody_orbit = np.loadtxt('8gyr/orbit', usecols=(1,2,3))


fig, ax = plt.subplots(ncols=1)
ax.set(
    title=r'Bound mass and stars',
    xlabel=r'X [pc]',
    ylabel=r'Y [pc]'
)

ax.plot(orbit[-200:,1], orbit[-200:,2], label='Orbit')
ax.scatter(nbody_orbit[-150:,1], nbody_orbit[-150:,2], label='NBODY', color='red')
ax.scatter(orbit[-1:,1], orbit[-1:,2], color='green', marker='+', s=1000, label='Final Position')

ax.legend()


```
